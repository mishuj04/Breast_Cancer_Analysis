# Observations
#### 1. The average tumor size is higher among people belonging to other races compared to White or Black individuals.  
#### 2. Divorced individuals tend to have larger tumor sizes.  
#### 3. People with undifferentiated tumors show larger tumor sizes.  
#### 4. Individuals with negative estrogen status have larger tumor sizes.  
#### 5. The average tumor size is higher in people who are alive compared to those who are deceased.  
#### 6. Married individuals show longer survival months on average.  
#### 7. People with well-differentiated tumors have longer survival months.  
#### 8. Individuals with negative estrogen status exhibit longer survival months.  
#### 9. A larger number of people are diagnosed with T2 stage breast cancer.  
#### 10. A larger number of people are diagnosed with N1 stage breast cancer.  
#### 11. Most patients are diagnosed with IIA (6th) stage breast cancer.  
#### 12. The majority of patients are diagnosed with Regional A stage breast cancer.  
#### 13. Grade 2 cancer is the most common among the patients.  

# Conclusions
#### 1. Race appears to play a role in tumor size variation, with non-White and non-Black groups showing higher tumor sizes.  
#### 2. Marital status may influence health outcomes, as divorced individuals show larger tumor sizes while married individuals tend to survive longer.  
#### 3. Tumor differentiation strongly impacts prognosis: undifferentiated tumors are linked with larger sizes, while well-differentiated tumors are linked with longer survival.  
#### 4. Estrogen receptor status significantly affects both tumor size and survival, with negative status linked to larger tumors but also longer survival.  
#### 5. Counterintuitively, people alive at the time of study had larger tumor sizes, suggesting survival depends on more than just tumor size (possibly treatment and other factors).  
#### 6. Certain cancer stages are more common, with T2, N1, IIA (6th), and Regional A stages being dominant in the dataset, indicating diagnostic clustering at these stages.  
#### 7. Grade 2 breast cancer is the most frequent, highlighting it as a common diagnosis level within the population.
